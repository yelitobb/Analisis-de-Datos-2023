```python
x = np.linspace(0, 15, 100)
y = x + np.random.randn(100)*0.5
```

```python
fig , ax = plt.subplots(figsize=(5,5))
gs = fig.add_gridspec(3,3)

ax1 = fig.add_subplot(gs[0,:])
ax1.plot(x, y, 'o', color = 'orange', ms = 5, lw=2 , mfc='lime', mec ='black', mew=2, ls =':', label = '$x^3$')
ax1.set_xticks([])
ax1.set_yticks([])

ax2 = fig.add_subplot(gs[1,0:2])
ax2.plot(x, y, '^-', color = 'red', ms = 5, lw=3, mfc='yellow', mec ='blue', mew=2, label = '$4x^2$')
ax2.set_xticks([])
ax2.set_yticks([])

ax3 = fig.add_subplot(gs[1:,2])
ax3.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$', alpha=0.3)
ax3.set_xticks([])
ax3.set_yticks([])

ax4 = fig.add_subplot(gs[2,0])
ax4.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$')
ax4.set_xticks([])
ax4.set_yticks([])

ax5 = fig.add_subplot(gs[2,1])
ax5.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$', alpha=0.3)
ax5.set_xticks([])
ax5.set_yticks([]);
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec


def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)

fig = plt.figure(constrained_layout=True)

gs = GridSpec(3, 3, figure=fig)
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y, 'o', color = 'orange', ms = 5, lw=2 , mfc='lime', mec ='black', mew=2, ls =':', label = '$x^3$')
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_xlim(0,30)
# identical to ax1 = plt.subplot(gs.new_subplotspec((0, 0), colspan=3))
ax2 = fig.add_subplot(gs[1, :-1])
ax2.plot(x, y, '^-', color = 'red', ms = 5, lw=3, mfc='yellow', mec ='blue', mew=2, label = '$4x^2$')
#ax2.set_xticks([])
#ax2.set_yticks([])
ax2.set_xlim(0,20)
ax3 = fig.add_subplot(gs[1:, -1])
ax3.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$', alpha=0.3)
ax3.set_xticks([])
ax3.set_yticks([])
ax4 = fig.add_subplot(gs[-1, 0])
ax4.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$')
ax4.set_xticks([])
ax4.set_yticks([])
ax5 = fig.add_subplot(gs[-1, -2])
ax5.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$', alpha=0.3)
ax5.set_xticks([])
ax5.set_yticks([]);

fig.suptitle("GridSpec")
#format_axes(fig)

plt.show()
```

```python
imagen = Image.open('Cat.jfif') 
imcat = np.array(imagen)
a,b,c = np.shape(imagen)
for i in range(0,a):
    for j in range(0,b):
        imcat[i][j][0] = imcat[i][j][2]
        imcat[i][j][1] = imcat[i][j][2]
        
imcat2 = np.array(imagen)
for i in range(0,a):
    for j in range(0,b):
        imcat2[i][j][0] = imcat2[i][j][1]
        imcat2[i][j][2] = imcat2[i][j][1]
        
imcat3 = np.array(imagen)
for i in range(0,a):
    for j in range(0,b):
        imcat3[i][j][1] = imcat3[i][j][0]
        imcat3[i][j][2] = imcat3[i][j][0]
        
imcat4 = np.array(imagen)
for i in range(0,a):
    for j in range(0,b):
        x = int(np.mean(imcat4[i][j]))
        imcat4[i][j][:] = x
        
        
fig,ax = plt.subplots(1,4,figsize=(15,15))
ax[0].imshow(imcat)
ax[1].imshow(imcat2)
ax[2].imshow(imcat3)
ax[3].imshow(imcat4)

```
